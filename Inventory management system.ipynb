```python
from tkinter import * 
from tkinter import ttk
import mysql.connector
from mysql.connector import Error
from tkinter import messagebox
from tkinter import filedialog

class my_app:
    def __init__(self,root):
        self.root=root
        self.root.geometry("1350x700+0+0")
        self.root.title("Product")
        bg_colour="brown"
        title=Label(self.root,text="Product Details",bd=10,bg=bg_colour,fg="white",font=("title new roman",30,"bold"),pady=2).pack(fill=X)
      
        #++++++++++veriabels+++++
        self.product_id=StringVar()
        self.product_name=StringVar()
        self.selling_price=IntVar()
        self.quantity=IntVar()
        
        
        
        
        
        
        #===========frames===========
        F1=LabelFrame(self.root,text="Inventory management",font=("times new roman",15,"bold"),fg="gold",bg=bg_colour)
        F1.place(x=5,y=80,width=250,height=620)
        
        pi=Label(F1,text="Product id:",font=("times new roman",16,"bold"),bg=bg_colour,fg="black").grid(row=0,column=0,padx=10,pady=10,sticky="w")
        self.text1=Entry(F1,width=10,font=("timesnew roman",12, "bold"),bd=5,relief=SUNKEN,textvariable=self.product_id).grid(row=1,column=0,padx=10,pady=10)                                                                               
                                                                                       
        pn=Label(F1,text="Product Name",font=("times new roman",16,"bold"),bg=bg_colour,fg="black").grid(row=2,column=0,padx=10,pady=10,sticky="w")
        self.text2=Entry(F1,width=10,font=("timesnew roman",16,"bold"),bd=5,relief=SUNKEN,textvariable=self.product_name).grid(row=3,column=0,padx=10,pady=10)                                                                               
        
        
        sp=Label(F1,text="Selling Price",font=("times new roman",16,"bold"),bg=bg_colour,fg="black").grid(row=4,column=0,padx=10,pady=10,sticky="w")
        self.text3=Entry(F1,width=10,font=("timesnew roman",16,"bold"),bd=5,relief=SUNKEN,textvariable=self.selling_price).grid(row=5,column=0,padx=10,pady=10)                                                                               
        
        ql=Label(F1,text="Quantity",font=("times new roman",16,"bold"),bg=bg_colour,fg="black").grid(row=6,column=0,padx=10,pady=10,sticky="w")
        self.text4=Entry(F1,width=10,font=("timesnew roman",16,"bold"),bd=5,relief=SUNKEN,textvariable=self.quantity).grid(row=7,column=0,padx=10,pady=10)    
        
        ins_b=Button(F1,command=self.insert,text="Insert",bg="cadetblue",fg="white",bd=1,pady=15,width=12,font="arial,12,bold").grid(row=8,column=0,padx=5,pady=5)
        shw_b=Button(F1,command=self.print_data,text="Show",bg="cadetblue",fg="white",bd=1,pady=15,width=12,font="arial,12,bold").grid(row=9,column=0,padx=5,pady=5)
        clr_b=Button(F1,command=self.clear,text="Clear",bg="cadetblue",fg="white",bd=1,pady=15,width=12,font="arial,12,bold").grid(row=10,column=0,padx=5,pady=5)
       
        
        
        F2=Frame(self.root)
        F2.place(x=250,y=80,width=1090,height=620)
        scr_y=ttk.Scrollbar(F2,orient=VERTICAL)
        self.area=Text(F2,yscrollcommand=scr_y.set)
        self.area=ttk.Treeview(F2,columns=('Product id','Name','Selling Price','Quantity'),yscrollcommand=scr_y.set)
        self.area.heading("Product id",text="Product Id")
        self.area.heading("Name",text="Name")
        self.area.heading("Selling Price",text="Selling Price")
        self.area.heading("Quantity",text="Quantity")
        self.area['show']='headings'
        self.area.column("Product id",width=150)
        self.area.column("Name",width=150)
        self.area.column("Selling Price",width=150)
        self.area.column("Quantity",width=150)
        scr_y.pack(side=RIGHT,fill=Y)
        scr_y.config(command=self.area.yview)
        self.area.pack(fill=BOTH,expand=1)
        
        
    def insert(self):
        id1=self.product_id.get()
        name=self.product_name.get()
        price=self.selling_price.get()
        quantity=self.quantity.get()
        
        connection=mysql.connector.connect(user="root",password="",host="localhost",database="inventory")
        if(id1=="" or name=="" or price<1 or quantity<1 ):
            messagebox.showerror("Error","please fill in all the required details")      
        else:
            try:
                cursor=connection.cursor()
                insert_query=(
                            """INSERT INTO inv(id,product_name,selling_price,quantity) VALUES(%s,%s,%s,%s)"""
                            )

                data=(id1,name,price,quantity)
                cursor.execute(insert_query,data)
                connection.commit()
                messagebox.showinfo("Success","Insert Data Susscessfull")
                self.product_id.set("")
                self.product_name.set("")
                self.selling_price.set(0)
                self.quantity.set(0)
         
        
            except Error as e:
                messagebox.showerror("Error"," Id already save ")
            
            
            
        
    def print_data(self):
        connection=mysql.connector.connect(user="root",password="",host="localhost",database="inventory")
        cursor=connection.cursor()
        My_Query="Select * from inv"
        cursor=connection.cursor()            
        cursor.execute(My_Query)
        result=cursor.fetchall()   
        for i in result:
             self.area.insert("",END,values=i)
                    
                    
                    
    def clear(self):
        for row in self.area.get_children():
            self.area.delete(row)
            self.product_id.set("")
            self.product_name.set("")
            self.selling_price.set(0)
            self.quantity.set(0)
                
        
                
root=Tk()
obj = my_app(root)
root.mainloop()
                
```
